# **T-test:**

Null Hpothesis H0: There is no difference between the mean of administered doses for the two states for September 2021 and Novemeber 2021.

Alternative Hypothesis H1: There is a difference between the mean of administered doses for the two states for September 2021 and Novemeber 2021.

T = d/(SD/âˆšn)

After applying T-test, if |T| > criticial_value, reject *H0*

Here critical_value = t(n-1,alp/2)

```python
#Creating difference column
df_md_septnov ["Difference"] = " "
```

```python
df_md_septnov["Difference"] = df_md_septnov['Administered_daily'] - df_la_septnov['Administered_daily']
```

```python
d = df_md_septnov['Difference'].mean()
n=len(df_md_septnov)
```

```python
stdd = df_md_septnov['Difference'].std()
print(stdd)
```

```python
denominator = stdd / math.sqrt(n)
```

```python
T = d / denominator
print(T)
```

```python
t_crit=2.003241
if (abs(T)>t_crit):
  print("We reject null hypothesis")
else:
  print("Accept null hypothesis")
```
